```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Applesauce1234',
    database='genz_data'
)

# Get meme data
df_memes = pd.read_sql("SELECT * FROM memes", conn)

print(df_memes)

# Add engagement score
df_memes['engagement'] = df_memes['likes'] + df_memes['shares']

# Find most viral meme
viral = df_memes.loc[df_memes['engagement'].idxmax()]
print("🔥 Most viral meme:", viral['meme_name'], "on", viral['platform'])

# Plot likes vs shares
df_memes.plot(kind='scatter', x='likes', y='shares')
plt.title("Memes: Likes vs Shares")
plt.xlabel("Likes")
plt.ylabel("Shares")
plt.show()
```

```python
# Get workout data
df_workouts = pd.read_sql("SELECT * FROM workouts", conn)

print(df_workouts)

# Bar chart: Calories burned per workout
df_workouts.plot(kind='bar', x='workout_name', y='calories_burned', legend=False)
plt.title("Calories Burned by Workout")
plt.ylabel("Calories")
plt.show()

# Mood breakdown
mood_counts = df_workouts['mood'].value_counts()
mood_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Workout Mood Breakdown")
plt.show()
```
